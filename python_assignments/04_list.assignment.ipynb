```python
# Question 6: Create a list of first 15 Fibonacci numbers

n_terms = 15
n1, n2 = 0, 1
count = 0

print("Fibonacci sequence:")
while count < n_terms:
    print(n1, end=" ")
    # Calculate the next term
    nth = n1 + n2
    # Update values for the next iteration
    n1 = n2
    n2 = nth
    count += 1
```

```python
# Question 10: Create a list of prime numbers between 1 and 50

prime_num = []

for i in range(2,51):
    is_prime = True
    for j in range(2, int(i ** 0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        prime_num.append(i)
        
print(f"List of prime numbers : {prime_num}")
```

```python
# Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]

nested_list = [1, [2, 3], [4, [5, 6]], 7]
stack = [(nested_list, 1)]
max_depth = 0

while stack:
    current_list, depth = stack.pop()
    max_depth = max(max_depth, depth)
    
    for i in current_list:
        if isinstance(i, list):
            stack.append((i, depth + 1))
            
print(f"Depth of nesting in {nested_list} : {max_depth}")
```
