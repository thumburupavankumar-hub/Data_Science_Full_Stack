```python
# Question 9: Count frequency of each character in string "hello"

string = "hello"
frequency = {}

for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
        
print(frequency)
```

```python
# Question 9: Count frequency of each character in string "hello"

string = "hello"
frequency = {}

for char in string:
    frequency[char] = frequency.get(char, 0) + 1
    
print(frequency)
```

```python
# Question 49: Create a dictionary with nested sets: {'groups': {{1, 2, 3}, {4, 5, 6}}, 'categories': {{'a', 'b'}, {'c', 'd'}}}

nested_set = {
    'groups': {frozenset({1, 2, 3}), frozenset({4, 5, 6})}, 
    'categories': {frozenset({'a', 'b'}), frozenset({'c', 'd'})}
}

print(f"Nested set : {nested_set}")
```

```python
# Question 50: Find union of all nested sets

nested_set = {
    'groups': {frozenset({1, 2, 3}), frozenset({4, 5, 6})}, 
    'categories': {frozenset({'a', 'b'}), frozenset({'c', 'd'})}
}

union_val = {}

for key, value in nested_set.items():
    union_val[key] = set().union(*value)
    
print(f"Union of all nested sets : {union_val}")
```
