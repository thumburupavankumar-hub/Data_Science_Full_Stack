```python
# Question 1: Create a set of first 10 natural numbers

natural_num = set(range(1,11))
print(natural_num)
```

```python
# Question 2: Add element 11 to set {1, 2, 3, 4, 5}

my_set = {1, 2, 3, 4, 5}
my_set.add(11)
print(my_set)
```

```python
# Question 3: Remove element 3 from set {1, 2, 3, 4, 5, 6}

my_set = {1, 2, 3, 4, 5, 6}

my_set.remove(3)

print(f"New set : {my_set}")
```

```python
# Question 4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}

my_set1 = {1, 2, 3, 4, 5}
my_set2 = {4, 5, 6, 7, 8}

inter_set = my_set1 & my_set2

# we can also use
# inter_set = my_set1.intersection(my_set2)

print(f"Intersection set : {inter_set}")
```

```python
# Question 5: Find the difference between {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}

my_set1 = {1, 2, 3, 4, 5}
my_set2 = {4, 5, 6, 7, 8}

difference = my_set1 - my_set2

print(f"Difference in {my_set1} and {my_set2} is {difference}.")
```

```python
# Question 6: Check if 5 is in set {1, 2, 3, 4, 6, 7, 8}

my_set = {1, 2, 3, 4, 6, 7, 8}

if 5 in my_set:
    print("5 is in my_set")
else:
    print("5 is not in my_set")
```

```python
# Question 7: Find the length of set {'a', 'b', 'c', 'd', 'e'}

my_set = {'a', 'b', 'c', 'd', 'e'}

print(f"Length of set : {len(my_set)}")
```

```python
# Question 8: Create a set of vowels from string "hello world"

char = "hello world"
vowels = "aeiou"
vowels_set = set()

for c in char:
    if c in vowels:
        vowels_set.add(c)
print(f"Vowels set : {vowels_set}")
        
```

```python
# Question 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set

my_list = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
new_list = set(my_list)

print(f"New list : {list(new_list)}")
```

```python
# Question 10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}

set1 = {1, 2, 3, 4, 5, 6}
subset1 = {1, 2, 3}

if subset1.issubset(set1):
    print(f"{subset1} is subset of {set1}")
else:
    print(f"{subset1} is not subset of {set1}")
```
