```python
# Question 1: Check if 25 is greater than 20 and less than 30

value = 25

if value > 20 and value < 30:
    print(True)
else:
    print(False)
```

```python
# Question 2: Check if string "python" is equal to "Python" (case sensitive)

str1 = "python"
str2 = "Python"

if str1 == str2:
    print(True)
else:
    print(False)
```

```python
# Question 3: Check if 15 is divisible by both 3 and 5

value = 15

result = (value % 3 == 0) and (value % 5 == 0)

print(f"Result : {result}")
```

```python
# Question 4: Check if 7 is not equal to 8

is_not_equal = 7 != 8

print(f"Result : {is_not_equal}")
```

```python
# Question 5: Check if 100 is greater than 50 or less than 25

if 100 > 50 or 100 < 25:
    print("True")
else:
    print("False")
```

```python
# Question 6: Check if 0 is falsy in Python

value = bool(0)

if value == True:
    print(True)
else:
    print(False)
```

```python
# Question 7: Check if empty string "" is falsy in Python

value = bool("")

if value == True:
    print(True)
else:
    print(False)
```

```python
# Question 8: Check if 42 is truthy in Python

value = bool(42)

if value == True:
    print(True)
else:
    print(False)
```

```python
# Question 9: Check if 10 is between 5 and 15 (inclusive)

value = 10

if value >= 5 and value <= 15:
    print(True)
else:
    print(False)
```

```python
# Question 9: Check if 10 is between 5 and 15 (inclusive)

value = 10 # we can also take user input
start_point = 5
end_point = 15

if value >= start_point and value <= end_point:
    print(True)
else:
    print(False)
```
