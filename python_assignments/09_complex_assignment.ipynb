```python
# Question 1: Create complex number 5 + 3j

complex_num = 5 + 3j

print(f"Complex number : {complex_num}")
```

```python
# Question 2: Find the real part of complex number 7 - 2j

complex_num = 7 - 2j

real_part = complex_num.real

print(f"Real part : {real_part}")
```

```python
# Question 3: Find the imaginary part of complex number 4 + 6j

complex_num = 4 + 6j

imaginary_part = complex_num.imag

print(f"Imaginary number : {imaginary_part}")
```

```python
# Question 4: Add two complex numbers: (3 + 2j) and (1 + 4j)

complex_num1 = (3 + 2j)
complex_num2 = (1 + 4j)

result = complex_num1 + complex_num2

print(f"Addition of two complex numbers : {result}")
```

```python
# Question 5: Multiply two complex numbers: (2 + 3j) and (1 + 2j)

complex_num1 = (2 + 3j)
complex_num2 = (1 + 2j)

result = complex_num1 * complex_num2

print(f"Multiplication of two complex numbers : {result}")
```

```python
# Question 6: Find the magnitude of complex number 6 + 8j

complex_num = 6 + 8j

magnitude = abs(complex_num)

print(f"Magnitude of {complex_num} : {magnitude}")
```

```python
# Question 7: Find the conjugate of complex number 5 - 7j

complex_num = 5 - 7j

conjugate = complex_num.conjugate()

print(f"Conjugate of {complex_num} : {conjugate}")
```

```python
# Question 8: Subtract complex numbers: (10 + 5j) - (3 + 2j)

subtract = (10 + 5j) - (3 + 2j)

print(f"Subtraction of complex numbers : {subtract}")
```

```python
# Question 9: Divide complex numbers: (8 + 6j) / (2 + 1j)

divide = (8 + 6j) / (2 + 1j)

print(f"Division of complex number : {divide}")
```

```python
# Question 10: Find the phase angle of complex number 1 + 1j

import math

complex_num = 1 + 1j

phase_angle = math.pi / 4

print(f"Phase angle : {phase_angle:.2f}")
```
