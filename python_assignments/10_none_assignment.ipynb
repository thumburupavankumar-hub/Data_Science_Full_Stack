```python
# Question 1: Create a variable and assign None to it

value = None

print(value)
```

```python
# Question 2: Check if None equals None

val1 = None
val2 = None

print(val1 == val2)
```

```python
# Question 3: Check if None is None

value = None

if value is None:
    print("None is None")
else:
    print("None is not None")
```

```python
# Question 4: Check if 0 is None

value = 0

if value is None:
    print(f"{value} is None")
else:
    print(f"{value} is not None")
```

```python
# Question 5: Check if empty string is None

string = ""

if string is None:
    print("Empty string is None")
else:
    print("Empty string is not None")
```

```python
# Question 6: Check if False is None

value = False

if value is None:
    print(f"{value} is None")
else:
    print(f"{value} is not None")
```

```python
# Question 7: Check if empty list is None

my_list = []

if my_list is None:
    print(f"{my_list} is None")
else:
    print(f"{my_list} is not None")
```

```python
# Question 8: Check if empty dictionary is None

my_dict = {}

if my_dict is None:
    print(f"{my_dict} is None")
else:
    print(f"{my_dict} is not None")
```

```python
# Question 9: Check if empty tuple is None

my_tuple = ()

if my_tuple is None:
    print(f"{my_tuple} is None")
else:
    print(f"{my_tuple} is not None")
```

```python
# Question 10: Check if empty set is None

my_set = set()

if my_set is None:
    print(f"{my_set} is None")
else:
    print(f"{my_set} is not None")
```
